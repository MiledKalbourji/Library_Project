```python
import pandas as pd  # import pandas library
import matplotlib.pyplot as plt  # import matplotlib library

# Load Excel file for each year 
df = pd.read_excel(r"Yearly Headcount.xlsx", sheet_name="Sheet1")

# Extract the data for each year
x_axis = ['2020', '2021', '2022', '2023']  # Years as x-axis labels
y_values = [df['Grand Total'][0], df['Grand Total'][1], df['Grand Total'][2], df['Grand Total'][3]]  #reading the grand total column

# Plotting each year as a single bar with unique colors
colors = ['blue', 'orange', 'green', 'red']  # distinct colors for each year
plt.bar(x_axis, y_values, color=colors) #plotting the x/y values given from the exel spread sheet

# Add data labels on top of each bar
for i, value in enumerate(y_values):
    plt.text(i, value, f'{int(value)}', ha='center', va='bottom')  # display value as an integer

# Labeling
plt.xlabel("Year")  # x-axis label
plt.ylabel("Total Amount of Students")  # y-axis label
plt.title("Yearly Total Student Headcount Comparison")  # title

plt.show()  # display the plot
```
