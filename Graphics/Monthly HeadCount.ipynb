```python
import pandas as pd #import pandas library
import matplotlib.pyplot as plt

df_2020 = pd.read_excel(r"Monthly Headcount.xlsx", sheet_name="2020") #loading excel file
df_2021 = pd.read_excel(r"Monthly Headcount.xlsx", sheet_name="2021") #loading excel file
df_2022 = pd.read_excel(r"Monthly Headcount.xlsx", sheet_name="2022") #loading excel file
df_2023 = pd.read_excel(r"Monthly Headcount.xlsx", sheet_name="2023") #loading excel file
df_2024 = pd.read_excel(r"Monthly Headcount.xlsx", sheet_name="2024") #loading excel file

x_axis = df_2020['Month'] #define which data will be represented on the x axis (months)
y_2020 = df_2020['Total'] #define which data will be represented on the y axis for 2020
y_2021 = df_2021['Total'] #define which data will be represented on the y axis for 2021
y_2022 = df_2022['Total'] #define which data will be represented on the y axis for 2022
y_2023 = df_2023['Total'] #define which data will be represented on the y axis for 2023
y_2024 = df_2024['Total'] #define which data will be represented on the y axis for 2024

#plot of each line for better spacing of the graph
plt.plot(x_axis, y_2020, label = '2020', linewidth=2) 
plt.plot(x_axis, y_2021, label = '2021', linewidth=2, linestyle='--') #dashed line
plt.plot(x_axis, y_2022, label = '2022', linewidth=2, linestyle='-.') #dashed dot line
plt.plot(x_axis, y_2023, label = '2023', linewidth=2, linestyle=':') #dotted line 
plt.plot(x_axis, y_2024, label = '2024', linewidth=2, linestyle=':') #dotted Line
plt.legend()

#Adding a y-axis label for clarity
plt.ylabel("Total amount of Students For a Month") # y-axis label

plt.show()
```
