```python
import pandas as pd  # Import pandas library for data manipulation
import matplotlib.pyplot as plt  # Import matplotlib library for creating visualizations

# Define the file path for the Excel file
file_path = r"Weekly Headcount.xlsx"

# Load the Excel sheet for 2020 into a DataFrame
df_2020 = pd.read_excel(file_path, sheet_name="2020")
# Load the Excel sheet for 2021 into a DataFrame
df_2021 = pd.read_excel(file_path, sheet_name="2021")
# Load the Excel sheet for 2022 into a DataFrame
df_2022 = pd.read_excel(file_path, sheet_name="2022")
# Load the Excel sheet for 2023 into a DataFrame
df_2023 = pd.read_excel(file_path, sheet_name="2023")
# Load the Excel sheet for 2024 into a DataFrame
df_2024 = pd.read_excel(file_path, sheet_name="2024")

# Clean column names for 2020 data by stripping extra spaces
df_2020.columns = df_2020.columns.str.strip()
# Clean column names for 2021 data by stripping extra spaces
df_2021.columns = df_2021.columns.str.strip()
# Clean column names for 2022 data by stripping extra spaces
df_2022.columns = df_2022.columns.str.strip()
# Clean column names for 2023 data by stripping extra spaces
df_2023.columns = df_2023.columns.str.strip()
# Clean column names for 2024 data by stripping extra spaces
df_2024.columns = df_2024.columns.str.strip()

# Define a function to create bar graphs for a given DataFrame and year
def plot_yearly_bar_graph(df, year):
    """
    Create a bar graph showing Best Week Totals and Worst Week Totals for a given year.
    
    Parameters:
    - df: DataFrame containing data for the year.
    - year: Year being plotted.
    """
    # Extract the months from the DataFrame
    months = df['Month']
    # Extract the Best Week Totals from the DataFrame
    best_totals = df['Grand Total of Best week']
    # Extract the Worst Week Totals from the DataFrame
    worst_totals = df['Grand Total of Worst Week']
    # Extract the Best Week dates from the DataFrame
    best_dates = df['Best Week']
    # Extract the Worst Week dates from the DataFrame
    worst_dates = df['Worst Week']

    # Generate the positions for the X-axis (each bar)
    x = range(len(months))  # One position per month
    # Set the width of each bar
    width = 0.4

    # Create a new figure for the bar graph
    plt.figure(figsize=(12, 6))  # Set the figure size to 12x6 inches

    # Create bars for the Best Week Totals
    best_bars = plt.bar(x, best_totals, width, label='Best Week Totals', color='blue')
    # Create bars for the Worst Week Totals, shifted to the right by the bar width
    worst_bars = plt.bar([i + width for i in x], worst_totals, width, label='Worst Week Totals', color='orange')

    # Loop through each bar in the Best Week Totals to add labels with the dates
    for bar, date in zip(best_bars, best_dates):
        plt.text(bar.get_x() + bar.get_width() / 2,  # X-coordinate of the text (centered on the bar)
                 bar.get_height() + 5,  # Y-coordinate of the text (slightly above the bar)
                 str(date),  # Text to display (Best Week date)
                 ha='center', va='bottom', fontsize=9, color='blue')  # Style and alignment of the text

    # Loop through each bar in the Worst Week Totals to add labels with the dates
    for bar, date in zip(worst_bars, worst_dates):
        plt.text(bar.get_x() + bar.get_width() / 2,  # X-coordinate of the text (centered on the bar)
                 bar.get_height() + 5,  # Y-coordinate of the text (slightly above the bar)
                 str(date),  # Text to display (Worst Week date)
                 ha='center', va='bottom', fontsize=9, color='orange')  # Style and alignment of the text

    # Label the X-axis as "Months"
    plt.xlabel('Months')
    # Label the Y-axis as "Total Count"
    plt.ylabel('Total Count')
    # Set the title of the graph to include the year
    plt.title(f'Yearly Headcount Analysis for {year}')
    # Set the X-axis ticks to be the months, centered between the two bars for each month
    plt.xticks([i + width / 2 for i in x], months, rotation=45)  # Rotate labels for better readability
    # Add a legend to differentiate Best and Worst Week Totals
    plt.legend()
    # Add a grid for the Y-axis for better visual clarity
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Automatically adjust the layout to prevent overlapping elements
    plt.tight_layout()
    # Display the graph
    plt.show()

# Call the function to plot the graph for 2020
plot_yearly_bar_graph(df_2020, 2020)
# Call the function to plot the graph for 2021
plot_yearly_bar_graph(df_2021, 2021)
# Call the function to plot the graph for 2022
plot_yearly_bar_graph(df_2022, 2022)
# Call the function to plot the graph for 2023
plot_yearly_bar_graph(df_2023, 2023)
# Call the function to plot the graph for 2024 
plot_yearly_bar_graph(df_2024, 2024)
```
